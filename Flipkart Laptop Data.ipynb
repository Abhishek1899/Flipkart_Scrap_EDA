```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
# extract the data for page number 1 only
page_number = 25
data = []

for i in tqdm(range(1, page_number)):

    # extract the page as a whole
    page = requests.get("https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&sort=popularity&page={}".format(i))

    # parse the whole page using a html parser
    soup = BeautifulSoup(page.content, 'html.parser')
    # soup

    # find_all() searches for the all the tags where the class is found and 
    # returns a list of all the occurences 
    all_laps = soup.findAll('a', href=True, attrs={'class':"_31qSD5"})

    # find the length of the elements extracted
    len(all_laps)

    for a in all_laps:
        
        processor = None
        ram = None
        os = None
        storage = None
        display = None

        name = a.find('div', attrs={'class':'_3wU53n'})
        ratings = a.find('div', attrs={'class':'hGSR34'})
        num_ratings = a.find('span', attrs={'class':'_38sUEc'})

        price = a.find(class_='_1vC4OE _2rQ-NK')
        org_price = a.find(class_='_3auQ3N _2GcJzG')


        allspecs = a.find('ul', attrs={'class':'vFw0gD'})
        sp = allspecs.findAll(class_="tVe95H")

        for i in range(len(sp)):
            if('Processor' in sp[i].text):
                processor = sp[i]

            elif('RAM' in sp[i].text):
                ram = sp[i]

            elif('Operating' in sp[i].text):
                os = sp[i]

            elif('SSD' in sp[i].text or 'HDD' in sp[i].text):
                storage = sp[i]

            elif('Display' in sp[i].text):
                display = sp[i]

        data.append({'name': name.text,
                     'ratings': ratings.text if ratings is not None else np.nan,
                     'num_ratings': num_ratings.text if num_ratings is not None else np.nan,
                    'processor': processor.text,
                    'ram': ram.text,
                    'os': os.text,
                    'storage': storage.text if storage is not None else np.nan,
                    'display': display.text,
                    'price': price.text,
                    'org_price': org_price.text if org_price is not None else np.nan})
```

```python
df = pd.DataFrame.from_dict(data) 

df 
```
